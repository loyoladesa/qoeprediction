# Método para separar o dataset em dois: Dias de Semana e Finais de Semana

```python
def obterDatasetFinalSemana(df):
  colunas = dados.columns
  df_final_semana = pd.DataFrame(columns = colunas)
  df_dias_semana =  pd.DataFrame(columns = colunas)
  

  numeroLinhas = len(df.index)
  linhasFinalSemana = []
  linhasDiaSemana = []
  
  #print('Numero de Linhas: ',numeroLinhas)
  numeroNovaLinhaFinalSemana = 0
  numeroNovaLinhaDiasSemana = 0

  for numeroLinha in range(numeroLinhas):
    #print('nº',numeroLinha)
    data = df.iloc[numeroLinha][0]
    date = datetime.strptime(data, '%Y-%m-%d %H:%M:%S').date()
    weekday = date.strftime("%A"), type(date.strftime("%A"))
    linha = df.iloc[numeroLinha]

    if (weekday[0] == "Saturday" or weekday[0] == 'Sunday'):
        df_final_semana.loc[numeroNovaLinhaFinalSemana]=linha
        numeroNovaLinhaFinalSemana = numeroNovaLinhaFinalSemana + 1
        linhasFinalSemana.append(linha)
        #print(numeroNovaLinha)
    else:
        df_dias_semana.loc[numeroNovaLinhaDiasSemana]=linha
        numeroNovaLinhaDiasSemana = numeroNovaLinhaDiasSemana + 1
        linhasDiaSemana.append(linha)
    
  return df_final_semana,df_dias_semana,linhasFinalSemana,linhasDiaSemana
```

```python
def obterDatasetDiasSemana(df):
  colunas = dados.columns
  df_dias_semana =  pd.DataFrame(columns = colunas)

  numeroLinhas = len(df.index)
  linhasDiaSemana = []
  #print('Numero de Linhas: ',numeroLinhas)
  
  numeroNovaLinhaDiasSemana = 0

  for numeroLinha in range(numeroLinhas):
    #print('nº',numeroLinha)
    data = df.iloc[numeroLinha][0]
    date = datetime.strptime(data, '%Y-%m-%d %H:%M:%S').date()
    weekday = date.strftime("%A"), type(date.strftime("%A"))
    linha = df.iloc[numeroLinha]

    if (not(weekday[0] == "Saturday" or weekday[0] == 'Sunday')):
        df_dias_semana.loc[numeroNovaLinhaDiasSemana]=linha
        numeroNovaLinhaDiasSemana = numeroNovaLinhaDiasSemana + 1
        linhasDiaSemana.append(linha)
   
        
  return df_dias_semana,linhasDiaSemana
```

```python
df_final_semana,df_dias_semana,linhasFinalSemana,linhasDiaSemana = obterDatasetFinalSemana(dados)
```
