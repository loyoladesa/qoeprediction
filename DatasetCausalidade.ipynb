## Causalidade

### Calcular Causalidade

```python
from statsmodels.tsa.stattools import grangercausalitytests
def calcularCausalidade(a,b):
  array = pd.DataFrame()
  array[0] = b
  array[1] = a
  array.dropna()
  try:
      #perform Granger-Causality test
      causalidade_test = grangercausalitytests(array, maxlag=[3])
      tupla = causalidade_test[3]
      testes = tupla[0]
      p_values = testes['params_ftest']
      if (p_values[1] < 0.05):
        return True
      return False
  except Exception as e:
      print("Exception: ",e)
      return False
```

### Stage 1

```python
dataset_throughput_rj = pd.read_csv(path_dataset_throughput_rj,sep=",")
dataset_throughput_sp = pd.read_csv(path_dataset_throughput_sp,sep=",")
dataset_packet_retransmits_sp = pd.read_csv(path_dataset_packet_retransmits_sp,sep=",")
dataset_packet_retransmits_rj = pd.read_csv(path_dataset_packet_retransmits_rj,sep=",")
```

```python
dataset_throughput_rj.dropna()
dataset_throughput_sp.dropna()
dataset_packet_retransmits_rj.dropna()
dataset_packet_retransmits_sp.dropna()
```

```python
alvo = dataset_packet_retransmits_rj[['sp']]
alvo
```

```python
S = []
for destino in pop_destino:
  S.append(dataset_throughput_sp[[destino]])
  S.append(dataset_packet_retransmits_rj[[destino]])
  S.append(dataset_packet_retransmits_sp[[destino]])
  if (destino != 'sp'):
    S.append(dataset_throughput_rj[[destino]])
```

```python
len(S)
```

```python
print(calcularCausalidade(S[0],S[20]))
```

```python
import networkx as nx
G = nx.DiGraph()

n = len(S)

for i in range(n):
  for j in range(n):
    print("i =",i,"j = ",j)
    array = pd.DataFrame()
    array[0] = S[i]
    array[1] = S[j] 
    array.dropna()
    if(i != j):
      existeCausalidade = calcularCausalidade(array[0],array[1])
      if (existeCausalidade):
          G.add_edge(i, j)
      existeCausalidade = calcularCausalidade(array[0],alvo)
      if (existeCausalidade):
          G.add_edge(i, 'C')
```

```python
print(G.nodes()) # [4, 5]

n_vertices = G.number_of_nodes()
n_arestas = G.number_of_edges()
print('vertices: ', n_vertices, '\narestas: ', n_arestas)

nx.draw(G, with_labels=True)
```

### Stage 2

```python
import networkx as nx
n = len(S)

for i in range(n):
  for j in range(n):
    print("i =",i,"j = ",j)
    array = pd.DataFrame()
    array[0] = S[i]
    array[1] = S[j] 
    array.dropna()
    if(i != j):
      existeCausalidade_i_c = calcularCausalidade(array[0],alvo)
      if (existeCausalidade_i_c):
          existeCausalidade_j_c = calcularCausalidade(array[1],alvo)
          if (existeCausalidade_j_c):
              existeCausalidade_i_j = calcularCausalidade(array[0],array[1])
              if (existeCausalidade_i_j):
                  G.remove_edge(i, 'C')
      
```

```python
print(G.nodes()) # [4, 5]

n_vertices = G.number_of_nodes()
n_arestas = G.number_of_edges()
print('vertices: ', n_vertices, '\narestas: ', n_arestas)
```
