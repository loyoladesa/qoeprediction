## Causalidade

### Calcular Causalidade

```python
from statsmodels.tsa.stattools import grangercausalitytests
def calcularCausalidade(a,b):
  array = pd.DataFrame()
  array[0] = b
  array[1] = a
  array.dropna()
  try:
      #perform Granger-Causality test
      causalidade_test = grangercausalitytests(array, maxlag=[3])
      tupla = causalidade_test[3]
      testes = tupla[0]
      p_values = testes['params_ftest']
      if (p_values[1] < 0.05):
        return True
      return False
  except Exception as e:
      print("Exception: ",e)
      return False
```

## Preliminary Graph Construction


### Stage 1

```python
dataset.dropna
```

```python
alvo = dataset[['qoe_value']]
alvo
```

```python
dataset.drop(columns=['qoe_value'], inplace=True)
```

```python
colunas = dataset.columns
print(len(colunas))
```

```python
S = []
for coluna in colunas:
  if (coluna != 'TimeStamp'):
    S.append(dataset[[coluna]])
```

```python
len(S)
```

```python
print(calcularCausalidade(S[0],S[20]))
```

```python
import networkx as nx
G = nx.DiGraph()

n = len(S)

for i in range(n):
  for j in range(n):
    print("i =",i,"j = ",j)
    array = pd.DataFrame()
    array[0] = S[i]
    array[1] = S[j]
    array.dropna()
    if(i != j):
      existeCausalidade = calcularCausalidade(array[0],array[1])
      if (existeCausalidade):
          G.add_edge(i, j)
      existeCausalidade = calcularCausalidade(array[0],alvo)
      if (existeCausalidade):
          G.add_edge(i, 'C')
```

```python
print(G.nodes()) # [4, 5]

n_vertices = G.number_of_nodes()
n_arestas = G.number_of_edges()
print('vertices: ', n_vertices, '\narestas: ', n_arestas)

nx.draw(G, with_labels=True)
```

### Stage 2

```python
import networkx as nx
n = len(S)

for i in range(n):
  for j in range(n):
    print("i =",i,"j = ",j)
    array = pd.DataFrame()
    array[0] = S[i]
    array[1] = S[j]
    array.dropna()
    if(i != j):
      existeCausalidade_i_c = calcularCausalidade(array[0],alvo)
      if (existeCausalidade_i_c):
          existeCausalidade_j_c = calcularCausalidade(array[1],alvo)
          if (existeCausalidade_j_c):
              existeCausalidade_i_j = calcularCausalidade(array[0],array[1])
              if (existeCausalidade_i_j):
                  # Check if the edge exists before removing
                  if G.has_edge(i, 'C'):
                        grau_entrada = G.in_degree('C')
                        print(f"Grau de entrada (in-degree) do vértice {'C'}: {grau_entrada}")
                        if grau_entrada > 1:
                          G.remove_edge(i, 'C')
```

```python
print(G.nodes()) # [4, 5]

n_vertices = G.number_of_nodes()
n_arestas = G.number_of_edges()
print('vertices: ', n_vertices, '\narestas: ', n_arestas)
```

```python
nx.draw(G, with_labels=True)
```
