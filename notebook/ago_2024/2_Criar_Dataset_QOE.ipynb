```python
def localizar_timestamp(df, timestamp_consultado):

    # Encontrar o Ã­ndice da linha onde o Timestamp Ã© imediatamente anterior
    index_anterior = df.loc[df['start'] < timestamp_consultado, 'start'].idxmax()
    # Obter a linha correspondente
    linha_anterior = df.loc[index_anterior]
    #print(linha_anterior)
    return index_anterior
```

```python
def concatena_datasets(df1,df2):
  if df2.loc[0]['start'] < df1.loc[0]['start']:
    df1, df2 = df2, df1

  print(df1.loc[0]['start'])
  print(df2.loc[0]['start'])

  df_concatenado = pd.DataFrame()
  # Iterando sobre as linhas usando .loc
  for i in df2.index:
      linha = df2.loc[i]
      start = linha['start']
      print(start)
      index = localizar_timestamp(df1, start)
      linha_df1 = df1.loc[[index]]
      linha_df2 = df2.loc[[i]]
      linha_junta = pd.concat([linha_df1.reset_index(drop=True), linha_df2.reset_index(drop=True)], axis=1)
      df_concatenado = pd.concat([df_concatenado, linha_junta], axis=0, ignore_index=True)
  return df_concatenado
```
