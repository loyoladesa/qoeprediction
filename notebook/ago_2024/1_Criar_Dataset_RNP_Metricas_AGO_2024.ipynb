### Percorre Datasets

Cria Dataset com as metricas do PoP de Origem RJ até os demais PoP da rede RNP.

```python
print(pastas)
print(pop_rnp)
```

```python
destino = 'rj'
for pasta in pastas:
  for pop in pop_rnp:
    if not(pasta == 'histogram-rtt'):
      path = f'{path_rnp}{pasta}/{pop}-{destino}.csv'
      #print(path)
      df_destino = pd.read_csv(path,sep=";")
      nome_coluna = pasta + "_" + pop + "_" + destino
      dataset_rnp[nome_coluna] = df_destino[['Valor']]
      #dataset_rnp[nome_coluna].fillna(value=0, inplace=True)
      mean_value = dataset_rnp[[nome_coluna]].mean()
      dataset_rnp[nome_coluna].fillna(value=mean_value[0], inplace=True)
    else:
      path = f'{path_rnp}{pasta}/{pop}-{destino}.csv'
      #print(path)
      df_destino = pd.read_csv(path,sep=";")
      nome_coluna = 'RTT_min' + "_" + pop + "_" + destino
      dataset_rnp[nome_coluna] = df_destino[['RTT_min']]
      #dataset_rnp[nome_coluna].fillna(value=0, inplace=True)
      mean_value = dataset_rnp[[nome_coluna]].mean()
      dataset_rnp[nome_coluna].fillna(value=mean_value[0], inplace=True)
      ############
      nome_coluna = 'RTT_avg' + "_" + pop + "_" + destino
      dataset_rnp[nome_coluna] = df_destino[['RTT_avg']]
      #dataset_rnp[nome_coluna].fillna(value=0, inplace=True)
      mean_value = dataset_rnp[[nome_coluna]].mean()
      dataset_rnp[nome_coluna].fillna(value=mean_value[0], inplace=True)
      ############
      nome_coluna = 'RTT_max' + "_" + pop + "_" + destino
      dataset_rnp[nome_coluna] = df_destino[['RTT_max']]
      #dataset_rnp[nome_coluna].fillna(value=0, inplace=True)
      mean_value = dataset_rnp[[nome_coluna]].mean()
      dataset_rnp[nome_coluna].fillna(value=mean_value[0], inplace=True)
```
