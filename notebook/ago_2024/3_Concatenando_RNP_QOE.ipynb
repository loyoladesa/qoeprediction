```python
def localizar_timestamp(df, timestamp_consultado):

    # Encontrar o índice da linha onde o Timestamp é imediatamente anterior
    index_anterior = df.loc[df['TimeStamp'] < timestamp_consultado, 'TimeStamp'].idxmax()
    # Obter a linha correspondente
    linha_anterior = df.loc[index_anterior]
    #print(linha_anterior)
    return index_anterior
```

```python
timestamp_inicio = '2024-08-10 18:19:08.307883'
timestamp_final = '2024-08-31 21:31:25.584911'
```

```python
index_inicio = localizar_timestamp(df_rnp, timestamp_inicio)
index_final = localizar_timestamp(df_rnp, timestamp_final)
```

```python
# Selecionar linhas entre os índices 1 e 3 (inclusive)
df_selecionado = df_rnp.loc[index_inicio:index_final]
```

```python
df_concatenado = pd.DataFrame()
```

```python
# Iterando sobre as linhas usando .loc
for i in df_qoe.index:
    linha = df_qoe.loc[i]
    start = linha['start']
    index = localizar_timestamp(df_selecionado, start)
    linha_df1 = df_selecionado.loc[[index]]
    linha_df2 = df_qoe.loc[[i]]
    linha_junta = pd.concat([linha_df1.reset_index(drop=True), linha_df2.reset_index(drop=True)], axis=1)
    df_concatenado = pd.concat([df_concatenado, linha_junta], axis=0, ignore_index=True)
```
